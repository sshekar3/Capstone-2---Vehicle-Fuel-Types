```python
#Create function to remove outliers that are 1.5* below or above the inter-quartile range

def removeOutliers(data, col):
    Q1 = np.quantile(data[col], 0.25)
    Q3 = np.quantile(data[col], 0.75)
    IQR = Q3 - Q1
 
    print("IQR value for column %s is: %s" % (col, IQR))
 
    lower_range = Q1 - 1.5 * IQR
    upper_range = Q3 + 1.5 * IQR
    outlier_free_list = [x for x in data[col] if (
        (x > lower_range) & (x < upper_range))]
    filtered_data = data.loc[data[col].isin(outlier_free_list)]
    return filtered_data
```

```python
#Remove outliers from the average gross income column and create a new dataframe we can use

df_income_trim = removeOutliers(df,'AGI_calc')
```

```python
a4_dims = (11.7, 8.27)
fig, ax = plt.subplots(figsize=a4_dims)
_ = sns.histplot(x = 'AGI_calc', data = df_income_trim, stat = 'density')

#Normalized AGI_calc with outliers removed; distribution looks better
```

```python
#Create subplots of average income vs. proportion of vehicle (fuel type)/all vehicles   for each fuel type


#Create a list of the different fuel types
fuels = df_income_trim.Fuel.unique()
#year_colors= {df_income_trim[((df_income_trim['Fuel'] == fuels[i]) & (df_income_trim['Year'] == '2019'))]:'r',
#              df_income_trim[((df_income_trim['Fuel'] == fuels[i]) & (df_income_trim['Year'] == '2020'))]:'g'}

fig,ax = plt.subplots(8,1, figsize= (15,25))
plt.tight_layout()
plt.subplots_adjust(hspace = 1)

#Create a dictionary to store fuel type: Pearson correlation pairs, p values (Average income vs. 
#                                          Ratio of vehicle(fuel type)/all vehicles)
fuel_inc_vr_corr = {}





for i in range(len(fuels)):

    
    x = df_income_trim[df_income_trim['Fuel'] == fuels[i]]['AGI_calc']
    y = df_income_trim[df_income_trim['Fuel'] == fuels[i]]['Vehicle Ratio']

    
    _ = sns.scatterplot(x = x, y = y, data = df_income_trim, hue = 'Year', palette = ['y','b'], ax = ax[i])
    _ =ax[i].set_title(f"{fuels[i]}")
    _ =ax[i].set_xlabel('AGI')
    _ =ax[i].set_ylabel('Vehicle Ratio')
    _ =ax[i].legend(loc = 'best')
    
    
    #Create best fit linear regression line
    m, b = np.polyfit(x, y, 1)
    _ = ax[i].plot(x, m*x + b, linestyle = '--', color = 'r')
    
    #Calculate pearson coefficient and p value for each subplot
    r,p = sp.stats.pearsonr(x,y)
    
    _ = ax[i].text(0.07, 0.4, f" r = {round(r,3)} p = {round(p,3)}", verticalalignment = 'center', horizontalalignment = 'center',transform = ax[i].transAxes)
    
    
    #Add fuel:correlation pairs to dictionary
    fuel_inc_vr_corr[fuels[i]] = [r,p]
    

    
    
#Battery Electric and Plug In Hybrid ownership appear to show a positive correlation with average income

    
```

```python
#Convert dictionary to DataFrame and sort to see correlations better 
fuel_corr_df = pd.DataFrame.from_dict(fuel_inc_vr_corr,orient = 'index')
fuel_corr_df.rename(columns = {0:'pearson_r',1:'p'}, inplace = True)
fuel_corr_df
```

```python
#Create subplots of pointplots to see vehicle ratio of each fuel type over time 


fig,ax = plt.subplots(8,1, figsize= (15,40))
plt.tight_layout()
plt.subplots_adjust(hspace = 1)

#Create a dictionary to store fuel type: Pearson correlation pairs (Average income vs. 
#                                          Ratio of vehicle(fuel type)/all vehicles)
fuel_corr_yr = {}

for i in range(len(fuels)):
    ##Create scatter plots to compare Year to Year vs. the Vehicle Ratio for each fuel type
       # To see if certain fuel type ownership changes over time
    
    x = df_income_trim[df_income_trim['Fuel'] == fuels[i]]['Year']
    y = df_income_trim[df_income_trim['Fuel'] == fuels[i]]['Vehicle Ratio']
    _ = sns.pointplot(x = x,y = y,ax = ax[i])
    _ =ax[i].set_title(f"{fuels[i]}")
    _ =ax[i].set_xlabel('Year')
    _ =ax[i].set_ylabel('Vehicle Ratio')
    #_ = ax[i].set_xticks(range(2019,2021))
    

    
    #Calculate pearson coefficient for each subplot
    r,p = sp.stats.pearsonr(x,y)
    _ = ax[i].text(0.07, 0.4, f" r = {round(r,3)} p = {round(p,3)}", verticalalignment = 'center', horizontalalignment = 'center',transform = ax[i].transAxes)
    #Add fuel:correlation pairs to dictionary
    fuel_corr_yr[fuels[i]] = [r,p]
    

    
    
#Battery Electric, Plug In Hybrid, Flex Fuel, Hybrid Gasoline ownership appear to show POSITIVE correlation with time
# Gasoline, Natural gas appear to show a NEGATIVE correlation with time

# P values for fuel types except Diesel and Hydrogen Fuel Cell are 0 indicating that there is a signficant change
# in vehicle ratio for each fuel type from 2019 to 2020
```
